# Attrition Analysis

```python
#What factors are associated with higher attrition rates?
sns.set(style="whitegrid")
factors = ['Age', 'MonthlyIncome', 'DistanceFromHome', 'JobSatisfaction', 'WorkLifeBalanceSatisfaction', 'JobInvolvement']

for factor in factors:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Attrition', y=factor, data=df)
    plt.title(f'Attrition vs {factor}')
    plt.show()
```

```python
#Are employees who work overtime more likely to leave the company?
plt.figure(figsize=(6, 4))
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title('Attrition vs Overtime')
plt.show()
```

```python
#How does job satisfaction relate to attrition?
plt.figure(figsize=(6, 4))
sns.countplot(x='JobSatisfaction', hue='Attrition', data=df)
plt.title('Attrition vs Job Satisfaction')
plt.show()
```

```python
#Does the department or job role impact attrition rates?
plt.figure(figsize=(8, 6))
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title('Attrition vs Department')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='JobRole', hue='Attrition', data=df)
plt.xticks(rotation=90)
plt.title('Attrition vs Job Role')
plt.show()
```

# Promotions and Career Progression

```python
#How long do employees typically stay in their current roles?
average_years_in_current_role = df['YearsInCurrentRole'].mean()
print(f"Average years in current role: {average_years_in_current_role:.2f} years")
```

```python
#How many years on average since the last promotion?
average_years_since_last_promotion = df['YearsSinceLastPromotion'].mean()
print(f"Average years since last promotion: {average_years_since_last_promotion:.2f} years")
```

```python
#Do certain job roles have higher rates of career progression?
plt.figure(figsize=(10, 6))
sns.barplot(x='YearsInCurrentRole', y='JobRole', data=df, ci=None)
plt.xlabel('Average Years in Current Role')
plt.ylabel('Job Role')
plt.title('Average Years in Current Role by Job Role')
plt.xticks(rotation=90)
plt.show()
```

# Management and Leadership

```python
#How long do employees typically stay with their current managers?
average_years_with_curr_manager = df['YearsWithCurrManager'].mean()
print(f"Average years with current manager: {average_years_with_curr_manager:.2f} years")
```

```python
#Are employees with more years of experience more likely to become managers?
plt.figure(figsize=(8, 6))
sns.scatterplot(x='YearsAtCompany', y='JobLevel', data=df)
plt.xlabel('Years at Company')
plt.ylabel('Job Level')
plt.title('Years at Company vs. Job Level')
plt.show()
```

# Training and Development

```python
#How many training times did employees have last year?
total_training_last_year = df['TrainingTimesLastYear'].sum()
print(f"Total training times last year: {total_training_last_year}")
```

```python
#Is there a correlation between training and performance?
correlation_matrix = df[['TrainingTimesLastYear', 'PerformanceRating']].corr()

plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation between Training and Performance')
plt.show()
```

```python
#Are employees with more training more likely to stay?
plt.figure(figsize=(6, 4))
sns.countplot(x='TrainingTimesLastYear', hue='Attrition', data=df)
plt.title('Attrition vs. Training Times Last Year')
plt.show()
```
